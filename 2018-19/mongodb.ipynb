```python
import pymongo
import time
```

```python
c = pymongo.MongoClient()
```

```python
b = c['inforet']['cran_tokens']
```

```python
m = {'$match': {
    'pos': 'VERB'
}}
g = {'$group': {
    '_id': {
        'partofspeech': '$pos', 'lemma': '$lemma'
    }, 'count': {'$sum': 1} 
}}
m1 = {'$match': {
    'count': {'$gte': 10}
}}
s = {'$sort': {
    'count': -1
}}
```

```python
for record in b.aggregate([g, m1, s]):
    print(record)
```
